The purpose of this analysis is to evaluate the quality of brake rotors produced during a one-week manufacturing run. By examining key measurements—thickness, diameter, surface roughness, and runout—we can identify which parts meet specifications, determine the yield for each measurement, and compare performance across production shifts. This information helps the quality control team detect process issues, reduce scrap, and maintain safety and reliability standards. Using Python and pandas, we will calculate statistics, assess yields, and provide actionable recommendations to improve production quality.

```python
import pandas as pd
```

```python
def calculate_basic_statistics(data: pd.DataFrame, measurement_column: str) -> dict:
    """
    Function to calculate basic statistics for a measurement column
    :param data: pandas DataFrame containing production measurements
    :param measurement_column: name of the column to analyze (e.g., 'Thickness_mm')
    :return: dictionary with keys 'mean', 'std', 'min', 'max' containing
             the corresponding statistical values
    """

    column = data[measurement_column]

    return {
        'mean': column.mean(),
        'std': column.std(),
        'min': column.min(),
        'max': column.max()
    }
```

```python
def count_parts_in_spec(data: pd.DataFrame, measurement_column: str, lower_limit: float, upper_limit: float):
    """
    Function to count the number of parts within specification limits
    :param data: pandas DataFrame containing production measurements
    :param measurement_column: name of the column to check (e.g., 'Thickness_mm')
    :param lower_limit: lower specification limit for the measurement
    :param upper_limit: upper specification limit for the measurement
    :return: integer count of values within the specified limits (inclusive)
    """
    
    column = data[measurement_column]

    series = (column >= lower_limit) & (column <= upper_limit)

    return series.sum()
```

```python
def analyze_by_shift(data: pd.DataFrame) -> pd.DataFrame:
    """
    Function to analyze production performance by work shift
    :param data: pandas DataFrame containing production measurements and a 'Shift' column
    :return: pandas DataFrame indexed by shift with average values for each measurement column
             ('Thickness_mm', 'Diameter_mm', 'Surface_Roughness_um', 'Runout_mm') and a count
             of parts produced per shift
    """
    measurement_columns = ['Thickness_mm', 'Diameter_mm', 'Surface_Roughness_um', 'Runout_mm']

    d = data.groupby('Shift')

    means = d[measurement_columns].mean()

    counts = d.size().rename('Count')

    results = means.join(counts)

    return results
```

```python
data = pd.read_csv('E4_brake_rotor_production.csv')
calculate_basic_statistics(data, 'Thickness_mm')
```

```python
count_parts_in_spec(data, 'Thickness_mm', 11.95, 12.05)
```

```python
analyze_by_shift(data).round(3)
```
