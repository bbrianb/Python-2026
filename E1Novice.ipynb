```python
def calculate_available_area(roof_length, roof_width, edge_clearance):
    return (roof_length - edge_clearance * 2) * (roof_width - edge_clearance * 2)
```

```python
def calculate_panel_energy(panel_wattage, hours_sunlight):
    return panel_wattage * hours_sunlight / 1000
```

```python
def calculate_installation_cost(num_panels, cost_per_panel, installation_cost_per_panel, fixed_cost):
    return num_panels * (cost_per_panel + installation_cost_per_panel) + fixed_cost
```

```python
size = 'Small'
location = 'Phoenix, AZ'
length = 30
width = 20
clearance = 2.5
panel_area = 17.5
wattage = 400
hours = 6.5
cost = 350
installation = 175
fixed = 3500
```

```python
print(size, 'house in', location)
print('Roof dimensions:', length, 'ft x', width, 'ft')
print('Edge clearance:', clearance, 'ft')
print('Panel area:', panel_area, 'sq ft')
print('Panel wattage:', wattage, 'W')
print('Peak sun hours:', hours, 'hours')
print(f'Cost per panel: ${cost:,.2f}')
print(f'Installation per panel: ${installation:,.2f}')
print(f'Fixed costs: ${fixed:,.2f}')
print('Available roof area:', total_area:=calculate_available_area(length, width, clearance), 'sq ft')
print('Energy production per panel per day:', panel_production:=calculate_panel_energy(wattage, hours), 'kWh')
print('Number of panels:', panels:=int(total_area/panel_area))
print(f'Total cost: ${calculate_installation_cost(panels, cost, installation, fixed):,.2f}')
print('Total daily energy production:', daily:=panels*panel_production, 'kWh')
print(f'Total annual energy production: {daily*365:,} kWh')
```
